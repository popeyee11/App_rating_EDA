```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('apps.csv')
```

data that missing values

```python
df.head(2)
```

```python
#converting nulls to nan
df['Rating']=df['Rating'].isnull().replace(' ',np.nan)
df['Size']=df['Size'].isnull().replace(' ',np.nan)
df['Current Ver']=df['Current Ver'].isnull().replace(' ',np.nan)
df['Android Ver']=df['Android Ver'].isnull().replace(' ',np.nan)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head(2)
```

```python
df['Reviews'].dtype
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'])
```

```python
df['Last_up_Date']=df['Last Updated'].dt.day
df['Last_up_Month']=df['Last Updated'].dt.month
df['Last_up_Year']=df['Last Updated'].dt.year
```

```python
df.head(2)
```

```python
df.drop('Last Updated',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df['Installs']=df['Installs'].str.replace('+','')
```

```python
#Univariate Analysis
plt.figure(figsize=(12,7))
df['Category'].value_counts()
car=sns.countplot(x='Category',data=df)
car.bar_label(car.containers[0])
plt.xlabel('category name')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
df.head(2)
```

```python
df['Rating'].value_counts()
Rating_count=df.groupby('Rating').agg({'Rating':'count'})
plt.pie(Rating_count['Rating'],labels=Rating_count.index,autopct='%1.2f%%')
plt.title('Percentage Data of Ratings')
```
